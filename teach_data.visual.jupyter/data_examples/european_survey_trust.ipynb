## Process and set country as index

```python
df_raw['cntry'].unique()
```

```python
df = df_raw.copy()
df['cntry'] = df['cntry'].apply(lambda x: x.decode().strip(' '))
df.set_index('cntry', inplace=True)
df
```

## Preparing a figure for publication
* Correctness
* Appealing
* ... and a bit more complicated

```python
import matplotlib.pyplot as plt
fig = plt.figure()
```

```python
ax = fig.gca()
df.groupby('ppltrst')['ppltrst'].count().plot(kind='bar', color=['blue']*11, ax=ax)
ax.figure
```

```python
alpha = 0.8
ax.set_xlabel('Response $r$', fontsize=16, alpha=alpha) # Labels
ax.set_ylabel('Number of occurrences', fontsize=16, alpha=alpha)
ax.set_title("Most people can be trusted or you can't be too careful", fontsize=16, alpha=alpha)
ax.grid(False)
ax.set_frame_on(False)
xticks = list(range(0, 11)) # Ticks
ax.xaxis.set_ticks(xticks)
ax.set_xticklabels(xticks, fontsize=12, alpha=alpha)
v = ax.set_xlim((-0.5, 10.5))
```

```python
ax.figure
```

## Plotnine - A Grammar of Graphics for Python
  * Rules for forming plots - "grammar"
  * Less stiff and more modern look and feel
  * Rich feature set
  * Simple plots should stay simple, complicated plots should be easer to think and create

```python
# conda install -c conda-forge plotnine
import plotnine as pn
df.reset_index(level=0, inplace=True)
```

```python
# Select data and aestetics mapping + geometric (stats) +  
pn.ggplot(df) + pn.aes(x='cntry', y='ppltrst') +  pn.geom_boxplot()
```

## Mini-assignment
* Plot a histogram of ppltrst
* geom: pn.geom_bar()
* Set labels using pn.xlab/pn.ylab

```python
hist_trst = pn.ggplot(df) +\
      pn.aes(x = 'ppltrst')+\
      pn.geom_bar(alpha=0.7) +\
      pn.facet_wrap('cntry', ncol=5) +\
      pn.xlab('Trust') + pn.ylab('Count')
hist_trst
```

## Saving
